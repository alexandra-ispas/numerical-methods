```octave
function [centroids] = clustering_pc(points, NC)
%n e nr de dimensiuni
  [m n] = size(points);
  x = m/NC;
  cluster = zeros(NC, x);
  centroids = zeros(NC, n);
  
  
  for i = 1 : NC
      for j = 0 : x-1;
      %cluster i are pe coloane nr liniilor corespunzatoare
          cluster(i, j + 1) = i + j * NC
      endfor  
      
      for j = 1 : n 
%centroids are pe coloane media elementelor pe fiecare dimensiune
        centroids(i, j) = double( mean( points (cluster(i, :), j ) ) );
      endfor
  endfor
  
  prev = centroids;
  
  while 1
      grup = zeros(NC, 1);
      
      poz(1 : NC) = 1;
      
      %pentru fiecare punct
      for i = 1 : m
      
          minim = inf;
          %ia fiecare centroid
          for j = 1 : NC
          %calculeaza distanta de la punct la fiecare centroid
              dis = double( norm( points(i, :) -  centroids(j,:) ) );
          
              if dis < minim
                
                  minim = dis;
                  x = j;
        %punctul trebuie adaugat in grupul centroidului
              endif
            
            endfor
              
%dupa ce parcurge toti centroizii,
%pune linia punctului in cel mai apropiat
           grup( x, poz(x) ) = i ;
           poz(x)++;
      endfor

      for i = 1 : NC
          for j = 1 : n
             centroids(i, j) = double( mean( points( grup(i, 1 : poz(i) - 1 ), j) ) );
          endfor
      endfor
      %err = norm(centroids - prev);
      if prev == centroids
          return;
      endif
      
      prev = centroids;
    endwhile
    
endfunction
```
