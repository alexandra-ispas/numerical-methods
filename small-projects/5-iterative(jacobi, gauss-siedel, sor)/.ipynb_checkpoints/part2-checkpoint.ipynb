```octave
function [centroids] = clustering_pc(points, NC)
  [m n] = size(points)
  cluster = zeros(NC, n);
  centroids = zeros(NC, n);
  minim = inf;
  for i = 1:NC
      for j = 0 : NC-1;
          cluster(i, j+1) = i + j*NC;
      endfor  
      for j = 1:n 
      centroids(i, j) = mean(points(cluster(i, :), j));
      endfor
  endfor
  prev = centroids;
  while 1
      %grup = zeros(NC, m);
      poz(1:NC) = 1;
      for i = 1:m
          k =1;
          for j = 1:NC
              dis = norm(points(i, :) -  centroids(j,:));
              if dis < minim
                  minim = dis;
                  x = j;
              endif
           endfor
           grup(x, poz(x)) = i ;
           poz(x)++;
      endfor
      for i = 1:NC
          for j = 1:n
          if points(grup(i, 1:k-1),j) ~= 0
          centroids(i,j) = mean(points(grup(i, 1:poz(x)-1),j));
          endif
          endfor
      endfor
      err = norm(centroids - prev);
      
      if err < 1e-6
        return;
      endif
      
      prev = cetroids;
    endwhile
endfunction
```
